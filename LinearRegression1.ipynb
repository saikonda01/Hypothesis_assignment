```python
import pandas as pd
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
#Visualize the data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.title('Scatter Plot of Years of Experience vs Salary Hike')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
#spliting the data
```

```python
X_salary = df[['YearsExperience']]  # Independent variable
Y_salary = df['Salary']              # Dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_salary, X_test_salary, Y_train_salary, Y_test_salary = train_test_split(
    X_salary, Y_salary, test_size=0.2, random_state=42
)
```

```python
#model building
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
LR.fit(X_train_salary,Y_train_salary)
```

```python
Y_pred_salary = LR.predict(X_test_salary)
```

```python
# Evaluateing the model
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test_salary, Y_pred_salary))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test_salary, Y_pred_salary))
print('Root Mean Squared Error:', metrics.mean_squared_error(Y_test_salary, Y_pred_salary, squared=False))

```

```python
# visualizing
plt.scatter(X_test_salary, Y_test_salary)
plt.plot(X_test_salary, Y_pred_salary, color='red', linewidth=2)
plt.title('Regression Line for Years of Experience vs Salary Hike')
plt.xlabel('Years of Experience')
plt.ylabel('Salary Hike')
plt.show()
```
